# Step5: Clean column names

```python
National_Time_Series_df.columns = National_Time_Series_df.columns.str.replace("*", "", regex=False)
National_Time_Series_df.columns = National_Time_Series_df.columns.str.replace("\n", "", )
National_Time_Series_df.columns = National_Time_Series_df.columns.str.strip()
```

# Top 10 NHS Provider by Long wait %

```python
long_wait_cols = [
    "Over 52 and up to 65 weeks",
    "Over 65 and up to 78 weeks",
    "Over 78 and up to 104 weeks",
    "Over 104 weeks"
]

Provider_df["Long Wait %"] = (
    Provider_df[long_wait_cols].sum(axis=1) /
    Provider_df["Total Waiting List"]
) * 100
top10=Provider_df.sort_values("Long Wait %", ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
plt.barh(top10["Organisation Name"], top10["Long Wait %"])

plt.title("Top 10 NHS Providers by Long Wait %")
plt.xlabel("Long Wait %")
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

# NHS RTT Performance Dashboard

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

National_Time_Series_df["Week Ending"] = pd.to_datetime(
    National_Time_Series_df["Week Ending"]
)

fig = plt.figure(figsize=(16,10),constrained_layout=True )
fig.suptitle("NHS RTT Performance Dashboard", fontsize=18, fontweight='bold')

# 1. Waiting List Trend
ax1 = plt.subplot(2,2,1)
ax1.plot(
    National_Time_Series_df["Week Ending"],
    National_Time_Series_df["Total Waiting List"]/1_000_000
)
ax1.set_title("National Waiting List Trend")
ax1.set_ylabel("Patients (Millions)")
ax1.tick_params(axis='x', rotation=45)

# 2. RTT Performance
ax2 = plt.subplot(2,2,2)
ax2.plot(
    National_Time_Series_df["Week Ending"],
    National_Time_Series_df["% within 18 weeks"]
)
ax2.axhline(y=94, linestyle='--')
ax2.set_title("RTT Performance vs 94% Target")
ax2.set_ylabel("% within 18 weeks")
ax2.tick_params(axis='x', rotation=45)

# 3. Providers
ax3 = plt.subplot(2,2,3)
top10 = Provider_df.sort_values("Long Wait %", ascending=True).tail(10)

ax3.barh(top10["Organisation Name"], top10["Long Wait %"])
ax3.invert_yaxis()
ax3.set_title("Top 10 Providers by Long Wait %")

for i, v in enumerate(top10["Long Wait %"]):
    ax3.text(v + 0.2, i, f"{v:.1f}%", va='center')

# 4. Specialists
ax4 = plt.subplot(2,2,4)
top10_function = Speciality_df.sort_values("Long Wait %", ascending=True).tail(10)

ax4.barh(top10_function["Treatment Function Name"], top10_function["Long Wait %"])
ax4.invert_yaxis()
ax4.set_title("Top 10 Specialists by Long Wait %")

for i, v in enumerate(top10_function["Long Wait %"]):
    ax4.text(v + 0.2, i, f"{v:.1f}%", va='center')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig(
    "NHS_RTT_Dashboard.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```
